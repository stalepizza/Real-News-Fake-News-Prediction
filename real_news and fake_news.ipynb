```python
import re
```

```python
def wordopt(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove digits
    text = re.sub(r'\d+', '', text)

    # Remove new line characters
    text = re.sub(r'\n', ' ', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
def output_label(n):
    if n==0:
        return "Its Fake"
    elif n==1:
        return "Its Genuine"
```

```python
def manual_testing(news):
    import pandas as pd  # Ensure pandas is available in the function scope
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)  # Correct the column reference
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    
    # Predictions
    pred_lr = LR.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    
    return (
        f"\n\nLR Prediction: {output_label(pred_lr[0])}"
        f"\nGBC Prediction: {output_label(pred_gbc[0])}"
        f"\nRFC Prediction: {output_label(pred_rfc[0])}"
    )


```
